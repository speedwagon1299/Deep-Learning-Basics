```python
import pandas as pd
import matplotlib.pyplot as mpt
import numpy as np
```

```python
x = np.array([10,20,40,70,100])
y = np.array([1,4,7,12,15])
size = x.size
```

```python
def ini_coeff(x,y):
    m_x = np.mean(x)
    m_y = np.mean(y)
    SS_xy = np.sum(x*y) - m_x*m_y*size
    SS_xx = np.sum(x*x) - m_x*m_x*size
    w = SS_xy/SS_xx
    b = m_y - w*m_x
    return (w,b)
```

```python
def gradient(rate,itr):
    w,b = ini_coeff(x,y)
    for i in range(itr):
        w = w - (rate/size)*np.sum(x*abs(y-(w*x+b)))
        b = b - (rate/size)*np.sum(abs(y-(w*x+b)))
    return (w,b)
```

```python
def plot_line(x,y,w,b):
    mpt.scatter(x,y,color='r',marker='o')
    y_pred = w*x + b
    mpt.plot(x,y_pred,color='b')
    mpt.xlabel('x')
    mpt.ylabel('y')
```

```python
w,b = gradient(0.000001,100)
plot_line(x,y,w,b)
```
