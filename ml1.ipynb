```python
import numpy as np
import matplotlib.pyplot as mpt
```

```python
x = np.array([1,2,3,4,5,6,7,6,8])
y = np.array([0,1,2,3,4,5,6,7,9])
```

```python
def reg_coeff(x,y):
    n = np.size(x)
    m_x = np.mean(x)
    m_y = np.mean(y)
    SS_xy = np.sum(y*x) - n*m_y*m_x     #sum of cross deviations    
    SS_xx = np.sum(x*x) - n*m_x*m_x     #sum of squared deviations
    b_1 = SS_xy/SS_xx   #slope
    b_0 = m_y - b_1*m_x     #y intercept
    return [b_0,b_1]
```

```python
def cost(b):
    c = (1/(2*9)*np.sum((y-(b[1]*x + b[0]))**2))
    return c
```

```python
def gradient(rate,itr) :
    b = reg_coeff(x,y)
    for i in range(itr):
        b[0] = b[0] - (rate*(1/9)*np.sum(x*abs(y-(b[0]+b[1]*x))))
        b[1] = b[1] - (rate*(1/9)*np.sum(abs(y-(b[1]*x+b[0]))))
    return b
```

```python
def plot_line(x,y,b):
    mpt.scatter(x,y,color = 'm', marker = 'o', s = 3)
    y_pred = b[0] + x*b[1]
    mpt.plot(x,y_pred, color = 'g')
    mpt.xlabel('x')
    mpt.ylabel('y')
```

```python
b = gradient(0.0000001,100)
plot_line(x,y,b)
```
